```python
import glob
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot():  
    for subdir in os.scandir("../results"):
       
        all_files = glob.glob(os.path.join(subdir.path, "*.out"))
        dfs = (pd.read_csv(file, delim_whitespace=True) for file in all_files)
        data = pd.concat(dfs, ignore_index=True)
        data["p95"] = data["p95"].divide(1000.0)
        
        p = sns.lineplot(data=data, x=data.target, y=data.p95,
            #marker='o',
                         legend=False,
#                         err_style="bars"
                        )

    
    sns.set_style("darkgrid")
    plt.title("Some amazing Title", x=0.18, y=1.08)
    plt.text(0, 10.3, "95th percentile latency in ms")
    plt.ylabel("")
    

    plt.xlim([0, 55000])
    plt.ylim([0, 10])
    
    
    
    plt.show()
#     plt.savefig('task1.png', bbox_inches='tight')

plot()
```
